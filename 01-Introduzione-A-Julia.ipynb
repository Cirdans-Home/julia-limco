Un altro tipo di dato importante in Julia e nel calcolo numerico sono le matrici. Le matrici sono array bidimensionali che possono essere utilizzati per rappresentare dati tabulari, sistemi di equazioni lineari e molto altro. Ecco un esempio di come creare e manipolare una matrice in Julia:

```julia
# Creazione di una matrice 3x3
matrix = [1 2 3; 4 5 6; 7 8 9]
# Accesso a un elemento specifico (ad esempio, l'elemento nella seconda riga e terza colonna)
element = matrix[2, 3]  # Restituisce 6
# Calcolo della trasposta della matrice
transposed_matrix = transpose(matrix)
```

# **Cicli e controllo del flusso**

Julia supporta le strutture di controllo del flusso comuni come `if`, `for` e `while`. Ecco alcuni esempi:

```julia# Struttura if

else
    println("x è minore o uguale a 5")
end

# **Funzioni**

Julia permette di definire funzioni in modo "esteso" o in una forma compatta che ricalca la notazione $f(x) = \dots$.

```julia
# Definizione compatta (stile matematico)
f(x) = x^2 + 2x + 1

# Definizione estesa (utile per algoritmi più complessi)
function g(x, y)
    risultato = sqrt(x^2 + y^2)
    return risultato
end

# Testiamo le funzioni
println("f(3) = ", f(3))
println("La norma del vettore (3,4) è: ", g(3, 4))
```

Ecco un esempio di funzione che combina `if`, `else` e cicli.

```julia
function test_primo(n)
    if n <= 1
        return false
    end
    
    # Controlliamo i divisori fino alla radice quadrata di n
    for i in 2:Int(floor(sqrt(n)))
        if n % i == 0  # % è l'operatore modulo (resto della divisione)
            return false
        end
    end
    
    return true
end

# Troviamo i numeri primi tra 1 e 50
primi = [i for i in 1:50 if test_primo(i)]
println("Numeri primi trovati: $primi")

`````julia
function è_primo(n)
    if n <= 1
        return false
    end

    # Controlliamo i divisori fino alla radice quadrata di n
    for i in 2:Int(floor(sqrt(n)))
        if n % i == 0  # % è l'operatore modulo (resto della divisione)
            return false
        end
    end

    return true
end

# Troviamo i numeri primi tra 1 e 50
primi = [i for i in 1:50 if è_primo(i)]
println("Numeri primi trovati: $primi")
`````

# L'operatore `.`

Se hai una funzione $f(x)$ e vuoi applicarla a ogni elemento di un vettore `v`, non serve scrivere un ciclo for. Basta aggiungere un punto `.` prima della parentesi.

```julia
v = [0, π/2, π, 3π/2]

# Calcolare il seno di ogni elemento del vettore
# sin(v) darebbe errore perché sin accetta uno scalare
risultato = sin.(v)

println("Seno degli angoli: $risultato")

# Funziona anche con gli operatori
quadrati = v .^ 2  # Eleva al quadrato ogni elemento singolarmente
